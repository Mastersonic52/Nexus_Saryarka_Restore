# Протокол №0: Создание изоляционной капсулы (venv)»


Прежде чем вводить код, мы должны создать для нашей системы стерильную среду. В мире Python это называется **виртуальным окружением**.

> Чистый `venv` (Virtual Environment) — это «золотой стандарт» индустрии. Его использование учит гигиене разработки и избавляет от классического кошмара: *«А у меня на компьютере эта библиотека уже была установлена, почему у соседа не работает?»*.


### Почему это важно?
Представьте, что Python — это операционная система Нексуса, а библиотеки — это внешние модули. Если один студент установит версию модуля 1.0, а другой 2.0, система может войти в конфликт. `venv` создает копию Python прямо в папке вашего проекта, отделяя её от остального компьютера.

### Как развернуть капсулу в PyCharm 2026.1:

1.  **Создание проекта:** При создании нового проекта выберите **"Custom Environment"** -> **"Generate New"** -> **"Virtualenv"**.
2.  **Путь:** Убедитесь, что папка окружения называется `.venv` и находится внутри папки проекта.
3.  **Активация:** PyCharm 2026.1 сделает это автоматически. Вы увидите название `(.venv)` в нижней части окна терминала.

### Ручной запуск (если что-то пошло не так)

Если терминал «молчит» и не видит окружение, используйте команды ручного управления терминалом «Нексуса»:

